```python
import pandas as pd
final_df=pd.read_csv('final_df.csv')
```

```python
final_df.shape
```

```python
final_df.sample(5)
```

```python
final_df['date'] = pd.to_datetime(final_df['date'])
```

```python
time_series_df = (
    final_df
    .groupby('date')
    .sum(numeric_only=True)
    .reset_index()
)
```

TIME SERIES ANALYSIS (Bio Age(5-17)  V/s Date)

```python
import matplotlib.pyplot as plt

plt.figure()
plt.plot(time_series_df['date'], time_series_df['bio_age_5_17'])
plt.xlabel("Date")
plt.ylabel("Biometric (Age 517)")
plt.title("Time Series of Biometric Updates (Age 517)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

STATE WISE DISTRIBUTION

```python
state_dist = (
    final_df
    .groupby('state')
    .sum(numeric_only=True)
    .reset_index()
)
```

Sate-wise Biometric Distribution (Age<17)

```python
plt.figure()
plt.bar(state_dist['state'], state_dist['bio_age_5_17'])
plt.xticks(rotation=90)
plt.ylabel("Biometric (Age < 17)")
plt.title("State-wise Biometric Distribution (Age <17)")
plt.tight_layout()
plt.show()
```

Sate-wise Biometric Distribution (Age>=17)

```python
plt.figure()
plt.bar(state_dist['state'], state_dist['bio_age_17_greater'])
plt.xticks(rotation=90)
plt.ylabel("Biometric (Age 17)")
plt.title("State-wise Biometric Distribution (Age 17)")
plt.tight_layout()
plt.show()
```

Sate-wise Enrollment Distribution (Age 0-5)

```python
plt.figure()
plt.bar(state_dist['state'], state_dist['enr_age_0_5'])
plt.xticks(rotation=90)
plt.ylabel("Enrollment (Age 0-5)")
plt.title("State-wise Enrollment Distribution (Age 0-5)")
plt.tight_layout()
plt.show()
```

Sate-wise Enrollment Distribution (Age 5-17)

```python

plt.figure()
plt.bar(state_dist['state'], state_dist['enr_age_5_17'])
plt.xticks(rotation=90)
plt.ylabel("Enrollment (Age 5-17)")
plt.title("State-wise Enrollment Distribution (Age 5-17)")
plt.tight_layout()
plt.show()
```

Sate-wise Enrollment Distribution (Age 18+)

```python

plt.figure()
plt.bar(state_dist['state'], state_dist['enr_age_18_greater'])
plt.xticks(rotation=90)
plt.ylabel("Enrollment (Age 18+)")
plt.title("State-wise Enrollment Distribution (Age 18+)")
plt.tight_layout()
plt.show()
```

```python
top_states_enr_18_greater = state_dist.sort_values(
    'enr_age_18_greater', ascending=False
).head(10)
```

```python
top_states_enr_18_greater.values
```

```python
plt.figure()
plt.bar(
    top_states_enr_18_greater['state'],
    top_states_enr_18_greater['enr_age_18_greater']
)

plt.xlabel("State")
plt.ylabel("Enrolment (Age  18)")
plt.title("Top 10 States by Aadhaar Enrolment (Age  18)")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

SECTION FOR DRILLDOWN ON TOP 5 STATES (ON THE BASIS OF DISTRIC)(ENROLLMENT 5-17)

FOR UTTAR PRADESH

```python
up_enr_5_17_df= final_df[final_df['state'] == 'Uttar Pradesh']
```

```python
up_district_dist = (
    up_enr_5_17_df
    .groupby('district')
    .sum(numeric_only=True)
    .reset_index()
    .sort_values('enr_age_5_17',ascending=False)
).head(10)
```

```python
plt.figure()
plt.barh(up_district_dist['district'], up_district_dist['enr_age_5_17'])
plt.xlabel("Enrolments (Age <=17)")
plt.title("District-wise Enrolments in Uttar Pradesh")
plt.tight_layout()
plt.show()
```
